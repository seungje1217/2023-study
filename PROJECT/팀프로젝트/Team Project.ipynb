# 2021~2022 인구수 증감률 그래프 

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일에서 데이터 로드
pop_seoul_2021 = pd.read_csv("서울시 인구 (2021).csv", header=2, usecols=['동별(2)', '소계.1'])
pop_seoul_2022 = pd.read_csv("서울시 인구 (2022).csv", header=2, usecols=['동별(2)', '소계.1'])

# 컬럼 이름 변경
pop_seoul_2021.rename(
    columns={
        '동별(2)': '지역',
        '소계.1': '인구수_2021',
    },
    inplace=True
)

pop_seoul_2022.rename(
    columns={
        '동별(2)': '지역',
        '소계.1': '인구수_2022',
    },
    inplace=True
)
# '소계' 행 제외
pop_seoul_2021 = pop_seoul_2021[~pop_seoul_2021['지역'].str.contains('소계')]
pop_seoul_2022 = pop_seoul_2022[~pop_seoul_2022['지역'].str.contains('소계')]

# 두 파일 병합
merged_df = pd.merge(pop_seoul_2021, pop_seoul_2022, on='지역')

# 데이터 타입 변환: 숫자로 변환 가능한 값만 남기고 나머지는 NaN으로 처리
merged_df['인구수_2021'] = pd.to_numeric(merged_df['인구수_2021'], errors='coerce')
merged_df['인구수_2022'] = pd.to_numeric(merged_df['인구수_2022'], errors='coerce')

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 2021년과 2022년의 인구 수 그래프 생성
plt.scatter(merged_df['지역'], merged_df['인구수_2021'], label='2021', marker='o', color='blue')
plt.scatter(merged_df['지역'], merged_df['인구수_2022'], label='2022', marker='s', color='red')

# 2021년 대비 2022년 증감율 계산 및 표시
percent_change = ((merged_df['인구수_2022'] - merged_df['인구수_2021']) / merged_df['인구수_2021']) * 100
for i, txt in enumerate(percent_change):
    plt.annotate(f'{txt:.2f}%', (merged_df['지역'][i], merged_df['인구수_2022'][i]), color='black', fontsize=10, ha='center')

plt.rc('font', family='Malgun Gothic')
plt.title('2021과 2022 서울 자치구별 인구 수 및 증감율 비교')
plt.xlabel('지역')
plt.ylabel('인구 수')
plt.xticks(rotation=90, fontsize=10)
plt.legend()
plt.grid(True)

plt.show()

# 증감율 출력
print("지역별 증감율:")
for index, row in merged_df.iterrows():
    region = row['지역']
    change_rate = ((row['인구수_2022'] - row['인구수_2021']) / row['인구수_2021']) * 100
    print(f"{region}: {change_rate:.2f}%")
```

# 2021 인구수 + 범죄 발생 

```python
import pandas as pd

# 인구 데이터 로드
pop_seoul = pd.read_csv("서울시 인구 (2021).csv", header=2, usecols=['동별(2)', '소계.1'])

# 컬럼명을 명확하게 변경
pop_seoul.rename(
    columns={
        pop_seoul.columns[0]: '자치구',
        pop_seoul.columns[1]: '인구수'
    },
    inplace=True
)

# 범죄 데이터 로드
crime_seoul = pd.read_csv('서울시 5대범죄 발생현황 (2021).csv', encoding='UTF-8', header=3)

# 첫 번째 열 제거
crime_seoul_without_first_column = crime_seoul.drop(crime_seoul.columns[0], axis=1)

# 컬럼명을 명확하게 변경
crime_seoul_without_first_column.rename(
    columns={
        crime_seoul_without_first_column.columns[0]: '자치구',
        crime_seoul_without_first_column.columns[1]: '범죄 발생 총계',
        crime_seoul_without_first_column.columns[2]: '범죄 검거 총계',
        crime_seoul_without_first_column.columns[3]: '살인 발생',
        crime_seoul_without_first_column.columns[4]: '살인 검거',
        crime_seoul_without_first_column.columns[5]: '강도 발생',
        crime_seoul_without_first_column.columns[6]: '강도 검거',
        crime_seoul_without_first_column.columns[7]: '강간·강제추행 발생',
        crime_seoul_without_first_column.columns[8]: '강간·강제추행 검거',
        crime_seoul_without_first_column.columns[9]: '절도 발생',
        crime_seoul_without_first_column.columns[10]: '절도 검거',
        crime_seoul_without_first_column.columns[11]: '폭력 발생',
        crime_seoul_without_first_column.columns[12]: '폭력 검거'
    },
    inplace=True
)

# '자치구'를 기준으로 중복된 행을 제거
pop_seoul.drop_duplicates(subset='자치구', keep='first', inplace=True)
crime_seoul_without_first_column.drop_duplicates(subset='자치구', keep='first', inplace=True)

# '자치구'를 기준으로 두 DataFrame을 병합
merged_data = pd.merge(pop_seoul, crime_seoul_without_first_column, on='자치구', how='inner')

merged_data.head(30)
```

```python
import pandas as pd

# 2021년 5대 범죄 발생 현황 데이터 로드
crime_seoul_2021 = pd.read_csv('서울시 5대범죄 발생현황 (2021).csv', encoding='UTF-8', header=3)

# '소계' 행 삭제
crime_seoul_2021 = crime_seoul_2021[~crime_seoul_2021.iloc[:, 0].str.contains('소계')]

# 1열 제거하여 새로운 데이터프레임 생성
crime_seoul_2021 = crime_seoul_2021.drop(crime_seoul_2021.columns[0], axis=1)

# 컬럼 이름 변경
crime_seoul_2021.rename(
    columns={
        crime_seoul_2021.columns[0]: '자치구',
        crime_seoul_2021.columns[1]: '발생 총계',
        crime_seoul_2021.columns[2]: '검거 총계',
        crime_seoul_2021.columns[3]: '살인 발생',
        crime_seoul_2021.columns[4]: '살인 검거',
        crime_seoul_2021.columns[5]: '강도 발생',
        crime_seoul_2021.columns[6]: '강도 검거',
        crime_seoul_2021.columns[7]: '강간·강제추행 발생',
        crime_seoul_2021.columns[8]: '강간·강제추행 검거',
        crime_seoul_2021.columns[9]: '절도 발생',
        crime_seoul_2021.columns[10]: '절도 검거',
        crime_seoul_2021.columns[11]: '폭력 발생',
        crime_seoul_2021.columns[12]: '폭력 검거'
    },
    inplace=True
)

# '소계' 행 삭제
crime_seoul_2021 = crime_seoul_2021[crime_seoul_2021['자치구'] != '소계']

# 2022년 5대 범죄 발생 현황 데이터 로드
crime_seoul_2022 = pd.read_csv('서울시 5대범죄 발생현황 (2022).csv', encoding='UTF-8', header=3)

# '소계' 행 삭제
crime_seoul_2022 = crime_seoul_2022[~crime_seoul_2022.iloc[:, 0].str.contains('소계')]

# 1열 제거하여 새로운 데이터프레임 생성
crime_seoul_2022 = crime_seoul_2022.drop(crime_seoul_2022.columns[0], axis=1)

# 컬럼 이름 변경
crime_seoul_2022.rename(
    columns={
        crime_seoul_2022.columns[0]: '자치구',
        crime_seoul_2022.columns[1]: '발생 총계',
        crime_seoul_2022.columns[2]: '검거 총계',
        crime_seoul_2022.columns[3]: '살인 발생',
        crime_seoul_2022.columns[4]: '살인 검거',
        crime_seoul_2022.columns[5]: '강도 발생',
        crime_seoul_2022.columns[6]: '강도 검거',
        crime_seoul_2022.columns[7]: '강간·강제추행 발생',
        crime_seoul_2022.columns[8]: '강간·강제추행 검거',
        crime_seoul_2022.columns[9]: '절도 발생',
        crime_seoul_2022.columns[10]: '절도 검거',
        crime_seoul_2022.columns[11]: '폭력 발생',
        crime_seoul_2022.columns[12]: '폭력 검거'
    },
    inplace=True
)

# '소계' 행 삭제
crime_seoul_2022 = crime_seoul_2022[crime_seoul_2022['자치구'] != '소계']

# 각 자치구별 총 범죄 발생 건수 계산
total_crime_2021 = crime_seoul_2021[['자치구', '발생 총계']].groupby('자치구').sum().reset_index()
total_crime_2022 = crime_seoul_2022[['자치구', '발생 총계']].groupby('자치구').sum().reset_index()

# 발생 건수를 %로 변환
crime_percent_change = pd.merge(total_crime_2021, total_crime_2022, on='자치구', suffixes=('_2021', '_2022'))
crime_percent_change['범죄 발생 증감률'] = ((crime_percent_change['발생 총계_2022'] - crime_percent_change['발생 총계_2021']) / crime_percent_change['발생 총계_2021']) * 100

# 필요한 칼럼만 선택
crime_percent_change = crime_percent_change[['자치구', '범죄 발생 증감률']]

# 각 범죄 카테고리(살인, 강도, 강간·강제추행, 절도, 폭력)별로 2021년과 2022년의 범죄 발생 증감률 계산
crime_categories = ['살인', '강도', '강간·강제추행', '절도', '폭력']
crime_category_percent_change = pd.DataFrame(columns=['자치구'])

for category in crime_categories:
    # 각 범죄 카테고리에 대한 발생 및 검거 데이터 추출
    columns_2021 = [f'{category} 발생', f'{category} 검거']
    columns_2022 = [f'{category} 발생', f'{category} 검거']
    
    # 각 범죄 카테고리에 대한 발생 증감률 계산
    percent_change = pd.merge(
        crime_seoul_2021[['자치구'] + columns_2021],
        crime_seoul_2022[['자치구'] + columns_2022],
        on='자치구',
        suffixes=('_2021', '_2022')
    )
    
    # 문자열 형식을 숫자로 변환
    percent_change[f'{category} 발생_2021'] = pd.to_numeric(percent_change[f'{category} 발생_2021'], errors='coerce')
    percent_change[f'{category} 발생_2022'] = pd.to_numeric(percent_change[f'{category} 발생_2022'], errors='coerce')
    
    # 각 범죄 카테고리에 대한 발생 증감률 계산
    percent_change[f'{category} 발생 증감률'] = ((percent_change[f'{category} 발생_2022'] - percent_change[f'{category} 발생_2021']) / percent_change[f'{category} 발생_2021']) * 100
    
    # 결과를 주요 데이터프레임에 추가
    crime_category_percent_change = pd.merge(
        crime_category_percent_change,
        percent_change[['자치구', f'{category} 발생 증감률']],
        on='자치구',
        how='outer'
    )

# 총 범죄에 대한 발생 및 검거 데이터 추출
columns_2021_total = ['발생 총계', '검거 총계']
columns_2022_total = ['발생 총계', '검거 총계']

# 총 범죄에 대한 발생 증감률 계산
total_percent_change = pd.merge(
    crime_seoul_2021[['자치구'] + columns_2021_total],
    crime_seoul_2022[['자치구'] + columns_2022_total],
    on='자치구',
    suffixes=('_2021', '_2022')
)

# 문자열 형식을 숫자로 변환
total_percent_change['발생 총계_2021'] = pd.to_numeric(total_percent_change['발생 총계_2021'], errors='coerce')
total_percent_change['발생 총계_2022'] = pd.to_numeric(total_percent_change['발생 총계_2022'], errors='coerce')

# 총 범죄에 대한 발생 증감률 계산
total_percent_change['총 범죄 발생 증감률'] = ((total_percent_change['발생 총계_2022'] - total_percent_change['발생 총계_2021']) / total_percent_change['발생 총계_2021']) * 100

# 결과를 주요 데이터프레임에 추가
crime_category_percent_change = pd.merge(
    crime_category_percent_change,
    total_percent_change[['자치구', '총 범죄 발생 증감률']],
    on='자치구',
    how='outer'
)
crime_category_percent_change.rename(
    columns={
        crime_category_percent_change.columns[0]:'자치구',
        crime_category_percent_change.columns[1]:'살인 발생',
        crime_category_percent_change.columns[2]:'강도 발생',
        crime_category_percent_change.columns[3]:'강간·강제추행 발생',
        crime_category_percent_change.columns[4]:'절도 발생',
        crime_category_percent_change.columns[5]:'폭력 발생',
        crime_category_percent_change.columns[6]:'총 범죄 발생 증감률',
        
    },
    inplace=True
)
# 결과를 출력
crime_category_percent_change.head(30)

# NaN은 2022년에 범죄가 아예 일어나지 않았던 경우 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 2021년 5대 범죄 발생 현황 데이터 로드
crime_seoul_2021 = pd.read_csv('서울시 5대범죄 발생현황 (2021).csv', encoding='UTF-8', header=3)

# '소계' 행 삭제
crime_seoul_2021 = crime_seoul_2021[~crime_seoul_2021.iloc[:, 0].str.contains('소계')]

# 1열 제거하여 새로운 데이터프레임 생성
crime_seoul_2021 = crime_seoul_2021.drop(crime_seoul_2021.columns[0], axis=1)

# 컬럼 이름 변경
crime_seoul_2021.rename(
    columns={
        crime_seoul_2021.columns[0]: '자치구',
        crime_seoul_2021.columns[1]: '발생 총계',
        crime_seoul_2021.columns[2]: '검거 총계',
        crime_seoul_2021.columns[3]: '살인 발생',
        crime_seoul_2021.columns[4]: '살인 검거',
        crime_seoul_2021.columns[5]: '강도 발생',
        crime_seoul_2021.columns[6]: '강도 검거',
        crime_seoul_2021.columns[7]: '강간·강제추행 발생',
        crime_seoul_2021.columns[8]: '강간·강제추행 검거',
        crime_seoul_2021.columns[9]: '절도 발생',
        crime_seoul_2021.columns[10]: '절도 검거',
        crime_seoul_2021.columns[11]: '폭력 발생',
        crime_seoul_2021.columns[12]: '폭력 검거'
    },
    inplace=True
)

# '소계' 행 삭제
crime_seoul_2021 = crime_seoul_2021[crime_seoul_2021['자치구'] != '소계']

# 2022년 5대 범죄 발생 현황 데이터 로드
crime_seoul_2022 = pd.read_csv('서울시 5대범죄 발생현황 (2022).csv', encoding='UTF-8', header=3)

# '소계' 행 삭제
crime_seoul_2022 = crime_seoul_2022[~crime_seoul_2022.iloc[:, 0].str.contains('소계')]

# 1열 제거하여 새로운 데이터프레임 생성
crime_seoul_2022 = crime_seoul_2022.drop(crime_seoul_2022.columns[0], axis=1)

# 컬럼 이름 변경
crime_seoul_2022.rename(
    columns={
        crime_seoul_2022.columns[0]: '자치구',
        crime_seoul_2022.columns[1]: '발생 총계',
        crime_seoul_2022.columns[2]: '검거 총계',
        crime_seoul_2022.columns[3]: '살인 발생',
        crime_seoul_2022.columns[4]: '살인 검거',
        crime_seoul_2022.columns[5]: '강도 발생',
        crime_seoul_2022.columns[6]: '강도 검거',
        crime_seoul_2022.columns[7]: '강간·강제추행 발생',
        crime_seoul_2022.columns[8]: '강간·강제추행 검거',
        crime_seoul_2022.columns[9]: '절도 발생',
        crime_seoul_2022.columns[10]: '절도 검거',
        crime_seoul_2022.columns[11]: '폭력 발생',
        crime_seoul_2022.columns[12]: '폭력 검거'
    },
    inplace=True
)

# '소계' 행 삭제
crime_seoul_2022 = crime_seoul_2022[crime_seoul_2022['자치구'] != '소계']

# 각 자치구별 총 범죄 검거 건수 계산
total_crime_2021 = crime_seoul_2021[['자치구', '검거 총계']].groupby('자치구').sum().reset_index()
total_crime_2022 = crime_seoul_2022[['자치구', '검거 총계']].groupby('자치구').sum().reset_index()

# 발생 건수를 %로 변환
crime_percent_change = pd.merge(total_crime_2021, total_crime_2022, on='자치구', suffixes=('_2021', '_2022'))
crime_percent_change['범죄 검거 증감률'] = ((crime_percent_change['검거 총계_2022'] - crime_percent_change['검거 총계_2021']) / crime_percent_change['검거 총계_2021']) * 100

# 필요한 칼럼만 선택
crime_percent_change = crime_percent_change[['자치구', '범죄 검거 증감률']]

# 각 범죄 카테고리(살인, 강도, 강간·강제추행, 절도, 폭력)별로 2021년과 2022년의 범죄 발생 증감률 계산
crime_categories = ['살인', '강도', '강간·강제추행', '절도', '폭력']
crime_category_percent_change = pd.DataFrame(columns=['자치구'])

for category in crime_categories:
    # 각 범죄 카테고리에 대한 발생 및 검거 데이터 추출
    columns_2021 = [f'{category} 발생', f'{category} 검거']
    columns_2022 = [f'{category} 발생', f'{category} 검거']
    
    # 각 범죄 카테고리에 대한 발생 증감률 계산
    percent_change = pd.merge(
        crime_seoul_2021[['자치구'] + columns_2021],
        crime_seoul_2022[['자치구'] + columns_2022],
        on='자치구',
        suffixes=('_2021', '_2022')
    )
    
    # 문자열 형식을 숫자로 변환
    percent_change[f'{category} 검거_2021'] = pd.to_numeric(percent_change[f'{category} 검거_2021'], errors='coerce')
    percent_change[f'{category} 검거_2022'] = pd.to_numeric(percent_change[f'{category} 검거_2022'], errors='coerce')
    
    # 각 범죄 카테고리에 대한 발생 증감률 계산
    percent_change[f'{category} 검거 증감률'] = ((percent_change[f'{category} 검거_2022'] - percent_change[f'{category} 검거_2021']) / percent_change[f'{category} 검거_2021']) * 100
    
    # 결과를 주요 데이터프레임에 추가
    crime_category_percent_change = pd.merge(
        crime_category_percent_change,
        percent_change[['자치구', f'{category} 검거 증감률']],
        on='자치구',
        how='outer'
    )

# 총 범죄에 대한 발생 및 검거 데이터 추출
columns_2021_total = ['발생 총계', '검거 총계']
columns_2022_total = ['발생 총계', '검거 총계']

# 총 범죄에 대한 발생 증감률 계산
total_percent_change = pd.merge(
    crime_seoul_2021[['자치구'] + columns_2021_total],
    crime_seoul_2022[['자치구'] + columns_2022_total],
    on='자치구',
    suffixes=('_2021', '_2022')
)

# 문자열 형식을 숫자로 변환
total_percent_change['검거 총계_2021'] = pd.to_numeric(total_percent_change['검거 총계_2021'], errors='coerce')
total_percent_change['검거 총계_2022'] = pd.to_numeric(total_percent_change['검거 총계_2022'], errors='coerce')

# 총 범죄에 대한 발생 증감률 계산
total_percent_change['총 범죄 검거 증감률'] = ((total_percent_change['검거 총계_2022'] - total_percent_change['검거 총계_2021']) / total_percent_change['검거 총계_2021']) * 100

# 결과를 주요 데이터프레임에 추가
crime_category_percent_change = pd.merge(
    crime_category_percent_change,
    total_percent_change[['자치구', '총 범죄 검거 증감률']],
    on='자치구',
    how='outer'
)
crime_category_percent_change.rename(
    columns={
        crime_category_percent_change.columns[0]:'자치구',
        crime_category_percent_change.columns[1]:'살인 검거',
        crime_category_percent_change.columns[2]:'강도 검거',
        crime_category_percent_change.columns[3]:'강간·강제추행 검거',
        crime_category_percent_change.columns[4]:'절도 검거',
        crime_category_percent_change.columns[5]:'폭력 검거',
        crime_category_percent_change.columns[6]:'총 범죄 검거 증감률',
        
    },
    inplace=True
)
# 결과를 출력
crime_category_percent_change.head(30)
```

# 인구수 발생에 따른 범죄 발생 연관관계

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일에서 데이터 로드
pop_seoul_2021 = pd.read_csv("서울시 인구 (2021).csv", header=2, usecols=['동별(2)', '소계.1'])
pop_seoul_2022 = pd.read_csv("서울시 인구 (2022).csv", header=2, usecols=['동별(2)', '소계.1'])

# 컬럼 이름 변경
pop_seoul_2021.rename(
    columns={
        '동별(2)': '지역',
        '소계.1': '인구수_2021',
    },
    inplace=True
)

pop_seoul_2022.rename(
    columns={
        '동별(2)': '지역',
        '소계.1': '인구수_2022',
    },
    inplace=True
)
# '소계' 행 제외
pop_seoul_2021 = pop_seoul_2021[~pop_seoul_2021['지역'].str.contains('소계')]
pop_seoul_2022 = pop_seoul_2022[~pop_seoul_2022['지역'].str.contains('소계')]

# 두 파일 병합
merged_df = pd.merge(pop_seoul_2021, pop_seoul_2022, on='지역')

# 데이터 타입 변환: 숫자로 변환 가능한 값만 남기고 나머지는 NaN으로 처리
merged_df['인구수_2021'] = pd.to_numeric(merged_df['인구수_2021'], errors='coerce')
merged_df['인구수_2022'] = pd.to_numeric(merged_df['인구수_2022'], errors='coerce')

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 2021년과 2022년의 인구 수 그래프 생성
plt.scatter(merged_df['지역'], merged_df['인구수_2021'], label='2021', marker='o', color='blue')
plt.scatter(merged_df['지역'], merged_df['인구수_2022'], label='2022', marker='s', color='red')

# 2021년 대비 2022년 증감율 계산 및 표시
percent_change = ((merged_df['인구수_2022'] - merged_df['인구수_2021']) / merged_df['인구수_2021']) * 100
for i, txt in enumerate(percent_change):
    plt.annotate(f'{txt:.2f}%', (merged_df['지역'][i], merged_df['인구수_2022'][i]), color='black', fontsize=10, ha='center')

plt.rc('font', family='Malgun Gothic')
plt.title('2021과 2022 서울 자치구별 인구 수 및 증감율 비교')
plt.xlabel('지역')
plt.ylabel('인구 수')
plt.xticks(rotation=90, fontsize=10)
plt.legend()
plt.grid(True)

plt.show()

# 2021년 5대 범죄 발생 현황 데이터 로드
crime_seoul_2021 = pd.read_csv('서울시 5대범죄 발생현황 (2021).csv', encoding='UTF-8', header=3)

# '소계' 행 삭제
crime_seoul_2021 = crime_seoul_2021[~crime_seoul_2021.iloc[:, 0].str.contains('소계')]

# 1열 제거하여 새로운 데이터프레임 생성
crime_seoul_2021 = crime_seoul_2021.drop(crime_seoul_2021.columns[0], axis=1)

# 컬럼 이름 변경
crime_seoul_2021.rename(
    columns={
        crime_seoul_2021.columns[0]: '자치구',
        crime_seoul_2021.columns[1]: '발생 총계',
        crime_seoul_2021.columns[2]: '검거 총계',
        crime_seoul_2021.columns[3]: '살인 발생',
        crime_seoul_2021.columns[4]: '살인 검거',
        crime_seoul_2021.columns[5]: '강도 발생',
        crime_seoul_2021.columns[6]: '강도 검거',
        crime_seoul_2021.columns[7]: '강간·강제추행 발생',
        crime_seoul_2021.columns[8]: '강간·강제추행 검거',
        crime_seoul_2021.columns[9]: '절도 발생',
        crime_seoul_2021.columns[10]: '절도 검거',
        crime_seoul_2021.columns[11]: '폭력 발생',
        crime_seoul_2021.columns[12]: '폭력 검거'
    },
    inplace=True
)

# '소계' 행 삭제
crime_seoul_2021 = crime_seoul_2021[crime_seoul_2021['자치구'] != '소계']

# 2022년 5대 범죄 발생 현황 데이터 로드
crime_seoul_2022 = pd.read_csv('서울시 5대범죄 발생현황 (2022).csv', encoding='UTF-8', header=3)

# '소계' 행 삭제
crime_seoul_2022 = crime_seoul_2022[~crime_seoul_2022.iloc[:, 0].str.contains('소계')]

# 1열 제거하여 새로운 데이터프레임 생성
crime_seoul_2022 = crime_seoul_2022.drop(crime_seoul_2022.columns[0], axis=1)

# 컬럼 이름 변경
crime_seoul_2022.rename(
    columns={
        crime_seoul_2022.columns[0]: '자치구',
        crime_seoul_2022.columns[1]: '발생 총계',
        crime_seoul_2022.columns[2]: '검거 총계',
        crime_seoul_2022.columns[3]: '살인 발생',
        crime_seoul_2022.columns[4]: '살인 검거',
        crime_seoul_2022.columns[5]: '강도 발생',
        crime_seoul_2022.columns[6]: '강도 검거',
        crime_seoul_2022.columns[7]: '강간·강제추행 발생',
        crime_seoul_2022.columns[8]: '강간·강제추행 검거',
        crime_seoul_2022.columns[9]: '절도 발생',
        crime_seoul_2022.columns[10]: '절도 검거',
        crime_seoul_2022.columns[11]: '폭력 발생',
        crime_seoul_2022.columns[12]: '폭력 검거'
    },
    inplace=True
)

# '소계' 행 삭제
crime_seoul_2022 = crime_seoul_2022[crime_seoul_2022['자치구'] != '소계']

# 각 자치구별 총 범죄 발생 건수 계산
total_crime_2021 = crime_seoul_2021[['자치구', '발생 총계']].groupby('자치구').sum().reset_index()
total_crime_2022 = crime_seoul_2022[['자치구', '발생 총계']].groupby('자치구').sum().reset_index()

# 발생 건수를 %로 변환
crime_percent_change = pd.merge(total_crime_2021, total_crime_2022, on='자치구', suffixes=('_2021', '_2022'))
crime_percent_change['증감률'] = ((crime_percent_change['발생 총계_2022'] - crime_percent_change['발생 총계_2021']) / crime_percent_change['발생 총계_2021']) * 100

# 그래프 그리기
plt.figure(figsize=(12, 6))

# 2021년과 2022년의 범죄 발생 건수를 표시
plt.scatter(crime_percent_change['자치구'], crime_percent_change['발생 총계_2021'], label='2021년', color='blue', marker='o')
plt.scatter(crime_percent_change['자치구'], crime_percent_change['발생 총계_2022'], label='2022년', color='red', marker='s')

# 증감률 표시
for i, txt in enumerate(crime_percent_change['증감률']):
    plt.annotate(f'{txt:.2f}%', (crime_percent_change['자치구'][i], crime_percent_change['발생 총계_2022'][i]), color='black', fontsize=10, ha='center', va='bottom')

plt.title('2021년에서 2022년으로의 범죄 발생 증감률 (%)')
plt.xlabel('자치구')
plt.ylabel('발생 총계')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

# CCTV 증감률과 범죄 검거 증감률 상관관계

```python
import pandas as pd
import matplotlib.pyplot as plt

# CCTV 설치 현황 데이터 로드
cctv_seoul = pd.read_csv('서울시 자치구 연도별 방범용 CCTV 운영 현황.csv', encoding='EUC-KR', header=0, usecols=['구분', '2021년', '2022년'])
cctv_seoul.rename(
    columns={
        cctv_seoul.columns[0]: '자치구',
        cctv_seoul.columns[1]: '2021년',
        cctv_seoul.columns[2]: '2022년'
    },
    inplace=True
)

# '2021년'과 '2022년' 열의 값을 숫자로 변환 (쉼표(,) 제거)
cctv_seoul['2021년'] = cctv_seoul['2021년'].str.replace(',', '').astype(float)
cctv_seoul['2022년'] = cctv_seoul['2022년'].str.replace(',', '').astype(float)

# '자치구' 열을 문자열로 변환
cctv_seoul['자치구'] = cctv_seoul['자치구'].astype(str)

# 2021년과 2022년 CCTV 설치 현황 그래프
plt.figure(figsize=(10, 6))
plt.scatter(cctv_seoul['자치구'], cctv_seoul['2021년'], label='CCTV 2021', marker='o', color='blue')
plt.scatter(cctv_seoul['자치구'], cctv_seoul['2022년'], label='CCTV 2022', marker='s', color='red')


# 증감률 계산 및 표시
percent_change = ((cctv_seoul['2022년'] - cctv_seoul['2021년']) / cctv_seoul['2021년']) * 100
for i, txt in enumerate(percent_change):
    plt.annotate(f'{txt:.2f}%', (cctv_seoul['자치구'][i], cctv_seoul['2022년'][i]), color='black', fontsize=10, ha='center', va='bottom')

plt.rc('font', family='Malgun Gothic')
plt.title('2021과 2022 서울 자치구별 CCTV 설치 현황 및 증감률 비교')
plt.xlabel('자치구')
plt.ylabel('CCTV 설치 대수')
plt.xticks(rotation=90, fontsize=10)
plt.legend()
plt.grid(True)


# 2021년 5대 범죄 발생 현황 데이터 로드
crime_seoul_2021 = pd.read_csv('서울시 5대범죄 발생현황 (2021).csv', encoding='UTF-8', header=3)

# '소계' 행 삭제
crime_seoul_2021 = crime_seoul_2021[~crime_seoul_2021.iloc[:, 0].str.contains('소계')]

# 1열 제거하여 새로운 데이터프레임 생성
crime_seoul_2021 = crime_seoul_2021.drop(crime_seoul_2021.columns[0], axis=1)

# 컬럼 이름 변경
crime_seoul_2021.rename(
    columns={
        crime_seoul_2021.columns[0]: '자치구',
        crime_seoul_2021.columns[1]: '발생 총계',
        crime_seoul_2021.columns[2]: '검거 총계',
        crime_seoul_2021.columns[3]: '살인 발생',
        crime_seoul_2021.columns[4]: '살인 검거',
        crime_seoul_2021.columns[5]: '강도 발생',
        crime_seoul_2021.columns[6]: '강도 검거',
        crime_seoul_2021.columns[7]: '강간·강제추행 발생',
        crime_seoul_2021.columns[8]: '강간·강제추행 검거',
        crime_seoul_2021.columns[9]: '절도 발생',
        crime_seoul_2021.columns[10]: '절도 검거',
        crime_seoul_2021.columns[11]: '폭력 발생',
        crime_seoul_2021.columns[12]: '폭력 검거'
    },
    inplace=True
)

# '소계' 행 삭제
crime_seoul_2021 = crime_seoul_2021[crime_seoul_2021['자치구'] != '소계']

# 2022년 5대 범죄 발생 현황 데이터 로드
crime_seoul_2022 = pd.read_csv('서울시 5대범죄 발생현황 (2022).csv', encoding='UTF-8', header=3)

# '소계' 행 삭제
crime_seoul_2022 = crime_seoul_2022[~crime_seoul_2022.iloc[:, 0].str.contains('소계')]

# 1열 제거하여 새로운 데이터프레임 생성
crime_seoul_2022 = crime_seoul_2022.drop(crime_seoul_2022.columns[0], axis=1)

# 컬럼 이름 변경
crime_seoul_2022.rename(
    columns={
        crime_seoul_2022.columns[0]: '자치구',
        crime_seoul_2022.columns[1]: '발생 총계',
        crime_seoul_2022.columns[2]: '검거 총계',
        crime_seoul_2022.columns[3]: '살인 발생',
        crime_seoul_2022.columns[4]: '살인 검거',
        crime_seoul_2022.columns[5]: '강도 발생',
        crime_seoul_2022.columns[6]: '강도 검거',
        crime_seoul_2022.columns[7]: '강간·강제추행 발생',
        crime_seoul_2022.columns[8]: '강간·강제추행 검거',
        crime_seoul_2022.columns[9]: '절도 발생',
        crime_seoul_2022.columns[10]: '절도 검거',
        crime_seoul_2022.columns[11]: '폭력 발생',
        crime_seoul_2022.columns[12]: '폭력 검거'
    },
    inplace=True
)

# '소계' 행 삭제
crime_seoul_2022 = crime_seoul_2022[crime_seoul_2022['자치구'] != '소계']

# 각 자치구별 총 범죄 검거 건수 계산
total_crime_2021 = crime_seoul_2021[['자치구', '검거 총계']].groupby('자치구').sum().reset_index()
total_crime_2022 = crime_seoul_2022[['자치구', '검거 총계']].groupby('자치구').sum().reset_index()

# 발생 건수를 %로 변환
crime_percent_change = pd.merge(total_crime_2021, total_crime_2022, on='자치구', suffixes=('_2021', '_2022'))
crime_percent_change['증감률'] = ((crime_percent_change['검거 총계_2022'] - crime_percent_change['검거 총계_2021'])/ crime_percent_change['검거 총계_2021']) * 100

# 그래프 그리기
plt.figure(figsize=(12, 6))

# 2021년과 2022년의 범죄 발생 건수를 표시
plt.scatter(crime_percent_change['자치구'], crime_percent_change['검거 총계_2021'], label='2021년', color='blue', marker='o')
plt.scatter(crime_percent_change['자치구'], crime_percent_change['검거 총계_2022'], label='2022년', color='red', marker='s')

# 증감률 표시
for i, txt in enumerate(crime_percent_change['증감률']):
    plt.annotate(f'{txt:.2f}%', (crime_percent_change['자치구'][i], crime_percent_change['검거 총계_2022'][i]), color='black',
                 fontsize=10, ha='center', va='bottom')

plt.title('서울시 총 범죄 검거 증감률 (%)')
plt.xlabel('자치구')
plt.ylabel('발생 총계')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
