```python
import pandas as pd

# CSV 파일에서 데이터 로드
pop_korea = pd.read_csv("인구총조사 (2022).csv", encoding='EUC-KR')

# 컬럼 이름 변경
pop_korea.rename(
    columns={
        pop_korea.columns[0]:'행정구역별(읍면동)',
        pop_korea.columns[1]:'2022년'
    },
    inplace=True
)
pop_korea.head(22)
```

```python
# 2021년과 2022년의 인구 수 비교 그래프 + % 출력
import csv
import pandas as pd
import matplotlib.pyplot as plt

def load_csv(file_path):
    with open(file_path, encoding='EUC-KR') as f:
        data = csv.reader(f)
        header = next(data)  
        header[0] = header[0].replace('\ufeff', '') 
        rows = [row for row in data]
    return header, rows

header1, data1 = load_csv('인구총조사 (2021).csv')
header2, data2 = load_csv('인구총조사 (2022).csv')

common_column_name = '행정구역별(읍면동)'
column_to_compare_2021 = '2021'  
column_to_compare_2022 = '2022'  

# 두 데이터프레임 생성
df1 = pd.DataFrame(data1, columns=header1)
df2 = pd.DataFrame(data2, columns=header2)

# 두 파일 병합
merged_df = pd.merge(df1, df2, on=common_column_name, how='outer', suffixes=('_2021', '_2022'))

# 데이터 타입 변환: 숫자로 변환 가능한 값만 남기고 나머지는 NaN으로 처리
merged_df[column_to_compare_2021] = pd.to_numeric(merged_df[column_to_compare_2021], errors='coerce')
merged_df[column_to_compare_2022] = pd.to_numeric(merged_df[column_to_compare_2022], errors='coerce')

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 2021년과 2022년의 인구 수 그래프 생성
plt.scatter(merged_df[common_column_name], merged_df[column_to_compare_2021], label='2021', marker='o', color='blue')
plt.scatter(merged_df[common_column_name], merged_df[column_to_compare_2022], label='2022', marker='s', color='red')

# 2021년 대비 2022년 증감율 계산 및 표시
percent_change = ((merged_df[column_to_compare_2022] - merged_df[column_to_compare_2021]) / merged_df[column_to_compare_2021]) * 100
for i, txt in enumerate(percent_change):
    plt.annotate(f'{txt:.2f}%', (merged_df[common_column_name][i], merged_df[column_to_compare_2022][i]), color='black', fontsize=10, ha='center')

plt.rc('font', family='Malgun Gothic')
plt.title('2021과 2022 인구 수 및 증감율 비교')
plt.xlabel(common_column_name)
plt.ylabel('인구 수')
plt.xticks(rotation=90, fontsize=10)
plt.legend()
plt.grid(True)

plt.show()

# 증감율 출력
print("지역별 증감율:")
for index, row in merged_df.iterrows():
    region = row[common_column_name]
    change_rate = ((row[column_to_compare_2022] - row[column_to_compare_2021]) / row[column_to_compare_2021]) * 100
    print(f"{region}: {change_rate:.2f}%")

```
